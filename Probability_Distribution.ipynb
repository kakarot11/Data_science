```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## PMF of Uniform Distribution

```python
from scipy.stats import randint

x = 1 # change the values to 2, 3, 4, 5 and check
low = 1
high = 6

# the flavour of ice-cream selected by a customer is equally likely to be any one of the
# 5 available flavours.
drandint = randint.pmf(x, low, high)
print('When x = 1, the PMF is = ', drandint)
```

```python
# visualization

low = 1
high = 6
x = np.arange(1, 6, 1)

# plotting the PMF
y1 = randint.pmf(x, low, high)

plt.bar(x, y1)
plt.ylim(0, 1.0)

plt.show()
```

## CDF of Uniform Distribution

```python
x = 2
prandint = randint.cdf(x, low, high)
print('when x = 2, the CDF is = ', prandint)
```

```python
low = 1
high = 6
x = np.arange(1, 6, 1)

# Plotting the CDF

y2 = randint.cdf(x, low, high)
plt.bar(x, y2)
plt.xlabel('Possible outcomes, x', fontsize = 15)
plt.ylabel('P(X <= x)', fontsize = 15)
plt.title('The Discrete Uniform Distribution - CDF', fontsize = 20)
plt.show()
```

# Binomial Distribution

## PMF of Binomial Distribution

```python
'''
The prob that exactly 1 customer purchases one strawberry ice cream
in the next 5 customers if all the 5 transactions involve the purchase
of a single ice cream with prob 0.2.'''

from scipy.stats import binom

x = 1 # no. of successes
n = 5 # no. of Bernoulli trials
p = 0.2 # prob of success

dbinom = binom.pmf(x, n, p)
dbinom
```

```python
# Visualizing the PMF of the Binomial Distribution
n = 5
p = 0.2
x = np.arange(0, 6, 1)

# Plotting the PMF
y1 = binom.pmf(x, n, p)

plt.bar(x, y1)
plt.xticks(x)
plt.xlabel('No. of success x, in n trials', fontsize = 15)
plt.ylabel('P(X = x)', fontsize = 15)
plt.title('The Binomial Distribution - PMF', fontsize = 20)
plt.show()
```

## CDF of Binomial Distribution

```python
'''
the prob of at most 1 success occurs in 5 trials, ie, at most 1 customer in the next
5 customers purchases one strawberry ice cream, assuming that all the transactions 
involve the purchase of a single ice cream, with the prob of success being 0.2'''

at_most = binom.cdf(x, n, p)
at_most
```

```python
# Visualizing the CDF of the Binomial Distribution
n = 5
p = 0.2
x = np.arange(0, 6, 1)

# Plotting the CDF
y2 = binom.cdf(x, n, p)

plt.bar(x, y2)
plt.xticks(x)
plt.xlabel('No. of success x, in n trials', fontsize = 15)
plt.ylabel('P(X <= x)', fontsize = 15)
plt.title('The Binomial Distribution - CDF', fontsize = 20)
plt.show()
```

# Poisson Distribution

## PMF of the Poisson Distribution

```python
from scipy.stats import poisson

x = 12 # no. of successes
mu = 10 # mean = lambda

# the prob that exactly 12 customers arrive in the next one hour:
pmf = poisson.pmf(x, mu)
pmf
```

```python
# Visualizing the PMF graph for the Poisson Distribution when lambda = 10

# no. of success
x = np.arange(0, 30, 1)
# lambda = mean = 10
mu = 10

# plotting the PMF
y1 = poisson.pmf(x, mu)

plt.bar(x, y1, color = 'red')
plt.xlabel('No. of success, X', fontsize = 15)
plt.ylabel('P(X = x)', fontsize = 15)
plt.title('The Poisson Distribution - PMF', fontsize = 20)
plt.show()
```

## CDF of the Poisson Distribution

```python
# The prob that 12 or less customers arrive in the next one hour
cdf = poisson.cdf(x, mu)
cdf
```

```python
# The event of interest that we want: P(X >= 12) = P(X = 12) + (1 - P(X <= 11))
pmf + (1 - cdf)
```

```python
# Visualizing the CDF graph for the Poisson Distribution when lambda = 10

# no. of success
x = np.arange(0, 30, 1)
# lambda = mean = 10
mu = 10

# plotting the CDF
y2 = poisson.cdf(x, mu)

plt.bar(x, y2, color = 'blue')
plt.xlabel('No. of success, X', fontsize = 15)
plt.ylabel('P(X <= x)', fontsize = 15)
plt.title('The Poisson Distribution - CDF', fontsize = 20)
plt.show()
```

# Continuous Uniform Distribution (CUD)

## The PDF for Continuous Uniform Distribution

```python
'''
The probability of weight taking any value between 101.7g and 124.3g for a single
ice cream scoop can be determined as follows.'''

from scipy.stats import uniform

x = 105
low = 101.7
high = 124.3
diff = high - low

# The value of x can be any value in the interval [101.7, 124.3]
# As an example x is taken as 105 here.
dunif = uniform.pdf(x, loc = low, scale = diff)
dunif
```

### <i> This here is the PDF. </i>

```python
# Visualizing the PDF for the Continuous Uniform Distribution
low = 101.7
high = 124.3
diff = high - low
x = np.arange(101, 126, 0.1)

# Plotting the PDF
y1 = uniform.pdf(x, loc = low, scale = diff)
plt.plot(x, y1, color = 'green')
plt.xlabel('Possible Outcomes', fontsize = 15)
plt.ylabel('P(X = x)', fontsize = 15)
plt.title('The Continuous Uniform Distribution - PDF', fontsize = 20)
plt.show()
```

## CDF for Continuous Uniform Distribution

```python
punif = uniform.cdf(x, loc = low, scale = diff)
punif
```

```python
# Visualizing the CDF for the Continuous Uniform Distribution
low = 101.7
high = 124.3
diff = high - low
x = np.arange(101, 126, 0.1)

# Plotting the CDF
y2 = uniform.cdf(x, loc = low, scale = diff)
plt.plot(x, y2, color = 'magenta')
plt.xlabel('Possible Outcomes', fontsize = 15)
plt.ylabel('P(X <= x)', fontsize = 15)
plt.title('The Continuous Uniform Distribution - CDF', fontsize = 20)
plt.show()
```

# Normal Distribution

```python
from scipy.stats import norm

# Finding the CDP using the z-score
x = 84
mu = 95
sd = 11

z = (x - mu)/sd

pnorm_z = norm.cdf(z)
pnorm_z
```

```python
p_x_greater_than_117 = 1 - norm.cdf(117, loc = mu, scale = sd)
p_x_greater_than_117
```

```python
p_x_between_84_117 = norm.cdf(117, loc = mu, scale = sd) - norm.cdf(84, loc = mu, scale = sd)
p_x_between_84_117
```
